**[QUEST 04] 파이썬을 사용하여 간단한 계산기 프로그램 만들기**

사용자로부터 두 개의 정수와 연산자(+,-,*,/,**)를 입력받아 계산 결과를 출력하는 프로그램을 만들어주세요

**조건**

사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다
정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 다시 입력받기를 시도합니다.
나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.
사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.
math 모듈을 사용하여 제곱 연산(**)을 지원합니다.
계산기를 통해 계속 계산을 할 것인지 입력받습니다.


````python
import math  #math 모듈을 사용하여 제곱 연산(**)을 지원

def cal():
    while True:  #중지하지 않으면 계산기를 통해 계쏙 계산을 할 것인지 입력받습니다.
        try:
              num1 = int(input('첫번째 정수를 입력하세요:'))
              num2 = int(input('두번째 정수를 입력하세요:'))
              mark = input('연산자를 입력하세요: (+, -, *, /, **)')

              if mark == '+':
                result = num1 + num2
              elif mark == '-':
                result = num1 - num2
              elif mark == '*':
                result = num1 * num2
              elif mark == '/':
                if num2 == 0: #두번째 정수가 0인경우
                  raise ZeroDivisionError('두번째 정수가 0입니다.') #ZeroDivisionError를 처리
                result = num1 / num2
              elif mark == '**':
                result = num1 ** num2
              print(result) #결과출력

        except ValueError as e: #사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리
            print(f"ValueError: 잘못된 입력입니다. 정수를 입력해주세요. {e}") #정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력
        except ZeroDivisionError as e: #나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리
            print(f"ZeroDivisionError: 0이 아닌 다른 정수를 입력하세요 {e}") #적절한 오류 메시지를 출력
        except Exception as e:  #사용자가 지원하지 않는 연산자를 입력한 경우
            print(f"알 수 없는 오류 발생. 지정된 연산자 외에 다른 연산자를 입력하지 마세요. {e}") #적절한 오류 메시지를 출력

        keep = input('계산을 계속하시겠습니까? (y/n): ')
        if keep.lower() != 'y':
            print('계산을 종료합니다.')
            break

cal ()
````
