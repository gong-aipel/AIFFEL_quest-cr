```python
# 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        # 사각형 종류 선택
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        # 선택에 따른 안내 메시지 출력
        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        else:
            print('1, 2, 3 중에서 다시 입력해주세요.')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>> ').split(','))
        area = width * vertical
        result = f'직사각형의 넓이는 : {area}'
        return result

    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>> ').split(','))
        area = base * height
        result = f'평행사변형의 넓이는 : {area}'
        return result

    def trape(self):
        top, base, height = map(int, input('윗변, 밑변, 높이를 입력하세요. 예시 : 윗변,밑변,높이\n >>> ').split(','))
        area = (top + base) * height / 2  # 사다리꼴 공식 수정
        result = f'사다리꼴의 넓이는 : {area}'
        return result

# 객체 생성 & 사용
a = Square()  # 객체 생성
if a.square == 1:
    print(a.rect())
elif a.square == 2:
    print(a.par())
elif a.square == 3:
    print(a.trape())
```

```python
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!

class Kiosk:
    def __init__(self):
        # 메뉴와 가격 초기화
        self.menu = ['americano', 'latte', 'mocha', 'choco_latte']
        self.price = [2000, 3000, 3000, 3000]
        self.order = []  # 주문한 메뉴를 저장할 리스트
        self.total_price = 0  # 총 주문 금액

    # 메뉴 출력 메서드
    def menu_print(self):
        print("=" * 40)
        print(f"{'Menu':<20}{'Price':>10}")
        print("-" * 40)
        for i, (menu, price) in enumerate(zip(self.menu, self.price), start=1):
            print(f"{i}. {menu:<20}{price:>10}원")
        print("=" * 40)

    # 주문 메서드
    def menu_select(self):
        while True:
            self.menu_print()
            try:
                choice = int(input("메뉴 번호를 선택하세요 (0: 주문 완료): "))
                if choice == 0:
                    break
                elif 1 <= choice <= len(self.menu):
                    quantity = int(input(f"{self.menu[choice - 1]}의 수량을 입력하세요: "))
                    self.order.append((self.menu[choice - 1], quantity, self.price[choice - 1] * quantity))
                    self.total_price += self.price[choice - 1] * quantity
                    print(f"{self.menu[choice - 1]} {quantity}개가 추가되었습니다.")
                else:
                    print("잘못된 메뉴 번호입니다. 다시 입력해주세요.")
            except ValueError:
                print("숫자를 입력해주세요.")

    # 지불 메서드
    def pay(self):
        print(f"총 주문 금액: {self.total_price}원")
        print("결제 방식을 선택하세요:")
        print("1. 현금(cash)")
        print("2. 카드(card)")
        while True:
            method = input("결제 방식 (1 또는 2, 'cash' 또는 'card'): ").strip().lower()
            if method in ['1', 'cash']:
                print("직원을 호출하겠습니다.")
                break
            elif method in ['2', 'card']:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("잘못된 입력입니다. 다시 선택해주세요.")

    # 주문서 출력 메서드
    def table(self):
        print("\n" + "=" * 40)
        print(f"{'Ordered Items':<20}{'Qty':>5}{'Price':>10}")
        print("-" * 40)
        for menu, quantity, total in self.order:
            print(f"{menu:<20}{quantity:>5}{total:>10}원")
        print("=" * 40)
        print(f"{'Total':<25}{self.total_price:>10}원")
        print("=" * 40)
```

```python
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
