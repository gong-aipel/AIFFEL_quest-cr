```python
# 필수 라이브러리 설치
!pip install diffusers transformers accelerate safetensors --quiet

# 라이브러리 불러오기
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# 모델 불러오기
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

# 프롬프트 목록 (각각 이미지 한 장씩 생성할 거야)
prompts = [
    "A modern minimalist glass chapel with a triangular roof, floating on water under a clear blue sky, futuristic architecture, photorealistic",
    "A scenic mountain trail surrounded by silver grass in autumn, overlooking vast fields and distant hills, cloudy sky, natural landscape photography",
    "A tranquil garden pond with lotus leaves, surrounded by volcanic stones and trees, traditional Korean landscape, cloudy weather, serene mood",
    "A beautiful camellia garden path with red flowers in bloom, round trimmed bushes, stone path, peaceful atmosphere, blue sky",
    "A coastal landscape with black volcanic rocks, turquoise ocean waves, wind turbines in the distance, cloudy sky, Jeju Island vibe"
]

# 이미지 생성 및 출력
images = []

for i, prompt in enumerate(prompts):
    print(f"Generating image {i+1}...")
    image = pipe(prompt, guidance_scale=7.5).images[0]
    images.append(image)

# 이미지 시각화
plt.figure(figsize=(20, 10))

for i, image in enumerate(images):
    plt.subplot(1, 5, i+1)
    plt.imshow(image)
    plt.axis("off")
    plt.title(f"Image {i+1}")

plt.tight_layout()
plt.show()
```
