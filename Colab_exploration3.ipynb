```
# üìå ColabÏóêÏÑú ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
!pip install -q torch torchvision
!pip install -q matplotlib

# üìÅ Google DriveÏóêÏÑú Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìú
from google.colab import files
uploaded = files.upload()

from PIL import Image
import numpy as np
import cv2
import torch
import torchvision.models.segmentation as models
from torchvision import transforms
import matplotlib.pyplot as plt

# üîç ÏóÖÎ°úÎìúÌïú Ïù¥ÎØ∏ÏßÄ Ïó¥Í∏∞
uploaded_filename = list(uploaded.keys())[0]
original_img = Image.open(uploaded_filename).convert("RGB")
original_img_resized = original_img.resize((512, 512))
original_cv = cv2.cvtColor(np.array(original_img_resized), cv2.COLOR_RGB2BGR)

# üß† DeepLab Î™®Îç∏ Î°úÎî© (COCO pretrained)
model = models.deeplabv3_resnet101(pretrained=True).eval()

# üîß Ïù¥ÎØ∏ÏßÄ Ï†ÑÏ≤òÎ¶¨
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])
input_tensor = transform(original_img_resized).unsqueeze(0)

# üîç ÎßàÏä§ÌÅ¨ ÏòàÏ∏°
with torch.no_grad():
    output = model(input_tensor)['out'][0]
mask = output.argmax(0).byte().cpu().numpy()

# üë§ ÏÇ¨Îûå(class 15) ÎßàÏä§ÌÅ¨
person_mask = (mask == 15).astype(np.uint8)
mask_3ch = np.stack([person_mask]*3, axis=2)

# üéØ (1) Ïù∏Î¨ºÎ™®Îìú: Î∞∞Í≤Ω ÏïÑÏõÉÌè¨Ïª§Ïã±
blurred_bg = cv2.GaussianBlur(original_cv, (25, 25), 0)
focused_img = np.where(mask_3ch == 1, original_cv, blurred_bg)

# üéØ (2) Î∞∞Í≤Ω Ìï©ÏÑ± (ÌÅ¨Î°úÎßàÌÇ§)
sky_background = np.full_like(original_cv, (255, 220, 200))  # Ïó∞Ìïú ÌïòÎäòÏÉâ
chroma_key_img = np.where(mask_3ch == 1, original_cv, sky_background)

# üéØ (3) Î¨∏Ï†ú ÏòÅÏó≠ ÏãúÍ∞ÅÌôî ÏòàÏãú
highlight_img = original_cv.copy()
highlight_img[person_mask == 0] = (highlight_img[person_mask == 0] * 0.5).astype(np.uint8)
cv2.circle(highlight_img, (130, 100), 20, (0, 0, 255), 2)  # Î™®ÏÑúÎ¶¨ Ïò§Î∂ÑÎ•ò
cv2.circle(highlight_img, (300, 400), 20, (255, 0, 0), 2)  # Î®∏Î¶¨ Í≤ΩÍ≥Ñ ÌùêÎ¶º

# üñºÔ∏è Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî
def show(title, img_bgr):
    plt.figure(figsize=(6,6))
    plt.title(title)
    plt.axis("off")
    plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
    plt.show()

show("ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ", original_cv)
show("1Ô∏è‚É£ Ïù∏Î¨ºÎ™®Îìú (ÏïÑÏõÉÌè¨Ïª§Ïã±)", focused_img)
show("2Ô∏è‚É£ ÌÅ¨Î°úÎßàÌÇ§ Î∞∞Í≤Ω Ìï©ÏÑ±", chroma_key_img)
show("3Ô∏è‚É£ ÏãúÎß®Ìã± ÏÑ∏Í∑∏Î©òÌÖåÏù¥ÏÖò Î¨∏Ï†ú ÏòÅÏó≠", highlight_img)
```
