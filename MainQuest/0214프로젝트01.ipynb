```python
# 데이터를 시각화하기 위한 함수
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(10,10))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
    # 문제1-1. 이미지에 나온 VGG16 모델을 구현하세요.
# [[YOUR CODE]]
# 이미지 전처리 및 데이터 생성기 (이미지 증강)
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 데이터 디렉토리 설정 (이미지 파일 경로 설정)
base_dir = "/content/drive/MyDrive/cats_and_dogs_filtered"

train_dir = os.path.join(base_dir, "train")  # 학습 데이터 경로
validation_dir = os.path.join(base_dir, "validation")  # 검증 데이터 경로

# 이미지 전처리 및 데이터 생성기 (이미지 증강)
train_datagen = ImageDataGenerator(
    rescale=1./255,  # 픽셀 값 정규화
    rotation_range=40,  # 이미지 회전
    width_shift_range=0.2,  # 좌우 이동
    height_shift_range=0.2,  # 상하 이동
    shear_range=0.2,  # 기울이기
    zoom_range=0.2,  # 확대/축소
    horizontal_flip=True,  # 수평 반전
    fill_mode='nearest'  # 결측값 채우기 방식
)

# 데이터 생성기
train_data_gen = train_datagen.flow_from_directory(
    train_dir,  # 학습 데이터 디렉토리
    target_size=(150, 150),  # 이미지 크기
    batch_size=20,  # 배치 크기
    class_mode='binary'  # 이진 분류 (고양이와 강아지)
train_datagen = ImageDataGenerator(
    rescale=1./255,  # 픽셀 값 정규화
    rotation_range=40,  # 이미지 회전
    width_shift_range=0.2,  # 이미지 좌우 이동
    height_shift_range=0.2,  # 이미지 상하 이동
    shear_range=0.2,  # 이미지 기울이기
    zoom_range=0.2,  # 이미지 확대
    horizontal_flip=True,  # 수평 반전
    fill_mode='nearest'  # 빈 공간 채우기
)

validation_datagen = ImageDataGenerator(rescale=1./255)  # 검증 데이터는 정규화만

# 데이터 생성기
train_data_gen = train_datagen.flow_from_directory(
    train_dir,  # 학습 데이터 디렉토리
    target_size=(150, 150),  # 이미지 크기
    batch_size=20,
    class_mode='binary'  # 이진 분류 (고양이/강아지)
)

validation_data_gen = validation_datagen.flow_from_directory(
    validation_dir,  # 검증 데이터 디렉토리
    target_size=(150, 150),  # 이미지 크기
    batch_size=20,
    class_mode='binary'  # 이진 분류
)
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16

# VGG16 모델을 가져오기 (ImageNet으로 미리 학습된 가중치를 사용)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))

# 모델을 고정시키고, 학습되지 않도록 설정
base_model.trainable = False

# 모델의 구조를 생성
model = models.Sequential([
    base_model,  # VGG16 모델을 추가
    layers.Flatten(),  # 출력층을 연결하기 위해 1차원으로 변환
    layers.Dense(512, activation='relu'),  # 완전 연결층 (512개 노드)
    layers.Dense(1, activation='sigmoid')  # 최종 출력층 (이진 분류: 고양이/강아지)
])

# 모델 컴파일
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # 이진 분류를 위한 손실 함수
              metrics=['accuracy'])

# 모델 요약 출력
model.summary()

# 모델 학습
history = model.fit(
    train_data_gen,
    steps_per_epoch=100,  # 학습 데이터를 한 에폭마다 100번 처리
    epochs=10,
    validation_data=validation_data_gen,
    validation_steps=50  # 검증 데이터를 50번 처리
)

# 학습된 모델 평가
test_loss, test_acc = model.evaluate(validation_data_gen)
print(f"Test accuracy: {test_acc}")
```
